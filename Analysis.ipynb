# Toronto Results

The revisions are as follows:
  * [s362ani-bug](https://github.com/QuLogic/specfem3d_globe/tree/s362ani-bug) is the HEAD
    of the s362ani-bug branch, which is devel + patch from Dimitri to use original s362ani code.
  * [Jan 2012](https://github.com/geodynamics/specfem3d_globe/commit/76e77ba9) is an old
    revision I found which appears to work.
  * [Last Good](https://github.com/geodynamics/specfem3d_globe/commit/b04f025d) is the last
    known good commit found during the bisect.
  * [First Bad](https://github.com/geodynamics/specfem3d_globe/commit/d1de5ba7) is the first
    known bad commit found during the bisect.
  * [v5.1.5](https://github.com/geodynamics/specfem3d_globe/commit/facdc6e7) is the last release.

Build types are:
  * ifort: 
        ifort (IFORT) 12.1.3 20120212
        Copyright (C) 1985-2012 Intel Corporation.  All rights reserved.
  * ifort + mcmodel=medium: Same as above with `-mcmodel=medium -shared-intel` added to
    `FLAGS_CHECK` and `CFLAGS` (and `FLAGS_NOCHECK` if applicable)
  * gfortran:
        GNU Fortran (GCC) 4.6.1
        Copyright (C) 2011 Free Software Foundation, Inc.
  * gfortran48:
        GNU Fortran (GCC) 4.8.1
        Copyright (C) 2013 Free Software Foundation, Inc.


```python
REVS = {
    'facdc6e7': 'v5.1.5',
    '76e77ba9': 'Jan 2012',
    'b04f025d': 'Last Good',
    'd1de5ba7': 'First Bad',
    '733e6b18': 's362ani-bug'
}
REV_BY_DATE = ['facdc6e7', '76e77ba9', 'b04f025d', 'd1de5ba7', '733e6b18'] # Sorted by date of commit
```

```python
ifort = {}
ifort_medium = {}
gfortran = {}
gfortran48 = {}
for r in REVS:
    ifort[r] = np.genfromtxt('Toronto/specfem3d_globe-%s/OUTPUT_FILES/AAE.IU.MXE.sem.ascii' % (r))
    ifort_medium[r] = np.genfromtxt('Toronto/specfem3d_globe-mcmodel-medium-%s/OUTPUT_FILES/AAE.IU.MXE.sem.ascii' % (r))
    gfortran[r] = np.genfromtxt('Toronto/specfem3d_globe-gfortran-%s/OUTPUT_FILES/AAE.IU.MXE.sem.ascii' % (r))
    gfortran48[r] = np.genfromtxt('Toronto/specfem3d_globe-gfortran48-%s/OUTPUT_FILES/AAE.IU.MXE.sem.ascii' % (r))
```

```python
for r in REV_BY_DATE:
    plt.figure(figsize=(16,12))

    ax1 = plt.subplot2grid((4,1), (0,0), rowspan=3)
    ax2 = plt.subplot2grid((4,1), (3,0), sharex=ax1)
    plt.setp(ax1.get_xticklabels(), visible=False)

    ax1.plot(ifort[r][:,0], ifort[r][:,1], label='ifort')
    ax1.plot(ifort_medium[r][:,0], ifort_medium[r][:,1], label='ifort + mcmodel=medium')
    ax1.plot(gfortran[r][:,0], gfortran[r][:,1], label='gfortran 4.6')
    ax1.plot(gfortran48[r][:,0], gfortran48[r][:,1], label='gfortran 4.8')
    ax1.legend(loc='best')
    ax1.set_ylabel('Displacement')
    ax1.set_title(REVS[r])

    ax2.plot(ifort[r][:,0], ifort[r][:,1] - ifort_medium[r][:,1], label='ifort vs. ifort+mcmodel=medium')
    ax2.plot(ifort[r][:,0], ifort[r][:,1] - gfortran[r][:,1], label='ifort vs. gfortran 4.6')
    ax2.plot(ifort_medium[r][:,0], ifort_medium[r][:,1] - gfortran[r][:,1], label='ifort+mcmodel=medium vs. gfortran 4.6')
    ax2.legend(loc='best')
    ax2.set_xlabel('Time')
    ax2.set_ylabel('Difference')
```

# Princeton Results

```python
PRINCETON_REV = '733e6b18'

ifort_princeton = np.genfromtxt('Princeton/Ebru_s_runs_are_NOT_OK_ifort/AAE.IU.MXE.sem.ascii')
ifort_medium_princeton = np.genfromtxt('Princeton/Ebru_s_runs_are_NOT_OK_ifort_medium/AAE.IU.MXE.sem.ascii')
gfortran_princeton = np.genfromtxt('Princeton/Ebru_s_runs_are_NOT_OK_gfortran/AAE.IU.MXE.sem.ascii')

plt.figure(figsize=(16,12))
ax1 = plt.subplot2grid((4,1), (0,0), rowspan=3)
ax2 = plt.subplot2grid((4,1), (3,0), sharex=ax1)
plt.setp(ax1.get_xticklabels(), visible=False)

ax1.plot(ifort_princeton[:,0], ifort_princeton[:,1], label='ifort')
ax1.plot(ifort_medium_princeton[:,0], ifort_medium_princeton[:,1], label='ifort + mcmodel=medium')
ax1.plot(gfortran_princeton[:,0], gfortran_princeton[:,1], label='gfortran')
ax1.legend(loc='best')
ax1.set_ylabel('Displacement')
ax1.set_title('s362ani-bug computed at Princeton')

ax2.plot(ifort_princeton[:,0], ifort_princeton[:,1] - ifort_medium_princeton[:,1], label='ifort vs. ifort+mcmodel=medium')
ax2.plot(ifort_princeton[:,0], ifort_princeton[:,1] - gfortran_princeton[:,1], label='ifort vs. gfortran')
ax2.plot(ifort_medium_princeton[:,0],
         ifort_medium_princeton[:,1] - gfortran_princeton[:,1], label='ifort+mcmodel=medium vs. gfortran')
ax2.legend(loc='best')
ax2.set_xlabel('Time')
ax2.set_ylabel('Difference')
```
